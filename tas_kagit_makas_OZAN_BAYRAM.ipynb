```python
import random
import time
def tas_kagit_makas_OZAN_BAYRAM():
 print("Taş, Kağıt, Makas Oyununa Hoşgeldiniz!")
 time.sleep(1)
 print("Kurallar:")
 time.sleep(1)
 print("1. Taş, makası yener.")
 time.sleep(1)
 print("2. Makas, kağıdı yener.")
 time.sleep(1)
 print("3. Kağıt, taşı yener.")
 time.sleep(1)
 print("İki tur kazanan oyunu kazanır.")
 time.sleep(1)
 print("Oyuna başlamadan önce önemli bir konu daha var!")
 time.sleep(1)

 def kelime_oyunu():
    print ("Söyle Dost Öyle Gir ! Bu kelime bir Lord of The Rings klasiğidir.")
    dogru_kelime ="Mellon"

    while True:
      kullanıcı_kelime=input("Lütfen bir kelime girin: ").strip()
      if kullanıcı_kelime:
        kullanıcı_kelime = kullanıcı_kelime[0].upper() + kullanıcı_kelime[1:].lower()
      if kullanıcı_kelime == dogru_kelime:
        print("Haydi Başlayalım.")
        break
      else:
        print(f"Kelime yanlış. İşte size ipucu: '{dogru_kelime}' yazmayı deneyebilirsin")
 kelime_oyunu()

 choices = ["taş", "kağıt", "makas", "çıkış"]
 while True:
  player_wins = 0
  computer_wins = 0

  while player_wins < 2 and computer_wins < 2 :
    player_choice = ""
    while player_choice not in choices:
      player_choice = input("Taş, kağıt veya makas seçin yada oyunu terketmek için çıkış seçin: ").lower()
      if player_choice not in choices:
        print("Geçersiz seçim. Lütfen tekrar deneyin.")
    if player_choice == "çıkış":
     print("Hoşçakal. Oyundan ayrıldınız.")
     return
    computer_choice = random.choice(choices)
    print(f"Bilgisayarın seçimi: {computer_choice}")

    if player_choice == computer_choice:
      print("Beraberlik")
    elif (player_choice == "taş" and computer_choice == "makas") or \
         (player_choice == "kağıt" and computer_choice == "taş") or \
         (player_choice == "makas" and computer_choice == "kağıt"):
        print ("Kazandınız!")
        player_wins += 1

    else:
      print("Bilgisayar Kazandı!")
      computer_wins += 1
    print(f"skor - Siz : {player_wins}, Bilgisayar: {computer_wins}")

  if player_wins == 2:
    print("Tebrikler! Oyunu kazandınız!")
  else:
    print("Bilgisayar oyunu kazandı!")
  while True:
    play_again = input("Başka bir oyun oynamak ister misiniz? (evet/hayır): ").lower()
    if play_again == "evet":
     print("Haydi yeniden oynayalım !")
     time.sleep(1)
     break
    elif play_again == "hayır":
     print("Hoşça kal")
     return
    else:
      print("Geçersiz seçim. Lütfen 'evet' yada 'hayır' yazınız")

tas_kagit_makas_OZAN_BAYRAM()
```
